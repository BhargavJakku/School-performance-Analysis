```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('C:\\Users\\Buncy\\Desktop\\Sem_1\\AIT_580\\FinalProject\\SchoolReport.csv')
data.head()
```

```python
print(data.shape)
x = data.dropna()
print(x.shape)
```

```python
df = x.groupby(['SCHOOL LEVEL*','OVERALL GRADE'])['OVERALL GRADE'].count()
df = df.to_frame()
df
```

```python
fig, ax = plt.subplots(figsize=(15,7))
df.unstack().plot(ax=ax)
```
